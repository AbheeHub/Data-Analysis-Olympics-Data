## Check all the data in the csv file and merge them

```python
athlete_events = pd.read_csv('athlete_events.csv')
```

```python
noc_regions = pd.read_csv('noc_regions.csv')
```

```python
athlete_events.head(10)
```

```python
noc_regions.head(10)
```

```python
olympics_merged = pd.merge(athlete_events,noc_regions,left_on = 'NOC',right_on = 'NOC')
```

```python
#medals = olympics_merged[olympics_merged['Medal'].isna() != True]['Medal']
highest_medals = olympics_merged.groupby('Team',sort = True)['Medal'].value_counts()
highest_medals.sort_values(ascending = False )[:10]
```

```python
highest_medals_year = olympics_merged.groupby('Team')['Year'].value_counts()
highest_medals_year.sort_values(ascending = False )[:10]
```

# Countries which won the highest number of gold medals


```python
highest_gold = olympics_merged[olympics_merged['Medal'] == 'Gold']['Team'].value_counts()
highest_gold[:20]
```

```python
# Person who won the most Gold medals in an olympic event
highest_gold = olympics_merged[olympics_merged['Medal'] == 'Gold']['Name'].value_counts()
highest_gold[:10]
```

# Athlete who won the most medals in one olympic event

```python
highest_person_medals_year = olympics_merged.groupby(['Name','Team','Games','Sport'])['Medal'].count()
highest_person_medals_year.sort_values(ascending = False )[:10]
```
