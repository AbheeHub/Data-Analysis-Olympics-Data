```python
def gold_medal(olympics_merged)  :
    if olympics_merged["Medal"] == "Gold" :
        return 1 
    else :
        return 0
def silver_medal(olympics_merged)  :
    if olympics_merged["Medal"] == "Silver" :
        return 1 
    else :
        return 0
def bronze_medal(olympics_merged)  :
    if olympics_merged["Medal"] == "Bronze" :
        return 1 
    else :
        return 0
    
olympics_merged["Gold"]   = olympics_merged.apply(lambda olympics_merged:gold_medal(olympics_merged),axis = 1) 
olympics_merged["Silver"] = olympics_merged.apply(lambda olympics_merged:silver_medal(olympics_merged),axis = 1) 
olympics_merged["Bronze"] = olympics_merged.apply(lambda olympics_merged:bronze_medal(olympics_merged),axis = 1) 
olympics_merged["Total Medals"] = olympics_merged["Gold"] + olympics_merged["Silver"] +olympics_merged["Bronze"]
```

```python
olympics_merged.head()
```

```python
total_medals_sport = olympics_merged.groupby('Sport')['Gold','Silver','Bronze','Total Medals'].sum()
total_medals_sport = total_medals_sport.sort_values(by = ['Total Medals','Gold','Silver','Bronze'],ascending = False)
total_medals_sport[:5]
```

# Change of Number of events in an olympics over the years

```python
olympics_merged.head()
```

```python
olympics_merged_summer = olympics_merged[olympics_merged['Season'] == 'Summer']
olympics_merged_winter = olympics_merged[olympics_merged['Season'] == 'Winter']
year = olympics_merged_summer['Year'].sort_values()
year = year.unique()
counts = []

for item in year:
    counter = list(olympics_merged_summer[olympics_merged_summer['Year'] == item]['Sport'].unique())
    counts.append(len(counter))
    
year1 = olympics_merged_winter['Year'].sort_values()
year1 = year1.unique()
counts1 = []

for item in year1:
    counter1 = list(olympics_merged_winter[olympics_merged_winter['Year'] == item]['Sport'].unique())
    counts1.append(len(counter1))        
```

```python
plt.figure(figsize=(15,13))

plt.subplot(211)
plt.xticks(rotation = 60)
sns.pointplot(year,counts)
plt.title("Sports over the last 12 decades in Summer Olympics")
plt.ylabel("Total Number of Sports")

plt.subplot(212)
plt.xticks(rotation = 60)
sns.pointplot(year1,counts1)
plt.title("Sports over the last 12 decades in Winter Olympics")
plt.ylabel("Total Number of Sports")
```
